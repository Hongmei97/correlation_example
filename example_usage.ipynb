```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from calculation_functions import calc_corr_dir, calc_corr_norm, calc_corr_glo, calc_corr_loc, calc_corr_gm, calc_corr_am, corr_to_1d
```

```python
# load the example data as 2d arrays
data = np.load('myxo_shear_example.npz')
x = data['x']
y = data['y']
u = data['u']
v = data['v']
xy_units = data['xy_units']
uv_units = data['uv_units']

# plot the field
plt.quiver(x, y, u, v)
plt.xlabel(f'x ({xy_units})')
plt.ylabel(f'y ({xy_units})')
plt.title('vector field')
plt.show()
```

```python
# calculate the correlation functions 2d arrays
corr_norm = calc_corr_norm(u, v)
corr_glo = calc_corr_glo(u, v)
corr_loc = calc_corr_loc(u, v)
corr_gm = calc_corr_gm(u, v)
corr_am = calc_corr_am(u, v)

# calculate the azimuthally averaged correlation function curves
dist_mean, corr_norm_mean = corr_to_1d(x, y, corr_norm)
_, corr_glo_mean = corr_to_1d(x, y, corr_glo)
_, corr_loc_mean = corr_to_1d(x, y, corr_loc)
_, corr_gm_mean = corr_to_1d(x, y, corr_gm)
_, corr_am_mean = corr_to_1d(x, y, corr_am)


# plot the correlation heatmap and correlation curve
plt.imshow(corr_norm)
plt.colorbar()
plt.xlabel(f'x ({xy_units})')
plt.ylabel(f'y ({xy_units})')
plt.title('correlation function')
plt.show()

plt.scatter(dist_mean, corr_norm_mean)
plt.xlabel(f'Distance ({xy_units})')
plt.ylabel('correlation function')
plt.title('correlation function curve')
plt.show()
```
